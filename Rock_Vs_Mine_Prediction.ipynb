## Logistic Regression Model is used because it works very well for binary classifier models

```python
# importing the Dependencies or Library
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Data Collection and Data Processing 

```python
# loading the dataset  
# If header files are not present in the dataset, we have to specify them as (header = None)
data = pd.read_csv(r"C:\Users\utkar\100 days ML\Rock Vs Mine Project\sonar data.csv",header = None)
```

```python
# first 5 data
data.head()
```

```python
# number of row and columns 
data.shape
```

```python
# checking for null values 
data.isnull().sum()
```

```python
# statistical definition about data  
data.describe()
```

```python
# total number of mine data and rock
data[60].value_counts()
```

```python
data.groupby(60).mean()
```

```python
# separating data and labels
x = data.drop(columns = 60 ,axis=1)
y = data[60]
```

```python
print(x)
```

```python
print(y)
```

## Training and Test data 

```python
X_train, X_test , Y_train , Y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
# checking the row and column of data 
print(x.shape, X_train.shape,X_test.shape)
```

```python
# printing the X_train and Y_train
print(X_train) # it is training data
print(Y_train) #it is training label
```

## Model Training 

```python
model = LogisticRegression()
```

```python
# training the logistic Regression model with training data 
model.fit(X_train,Y_train)
```

## Model Evaluation 

```python
# accuracy on training data most of the time its accuracy is more 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
# print the accuracy of training data 
print("Accuracy of training data is : " , training_data_accuracy)
```

```python
# accuracy on test data most of the time its accuracy is less then training 
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
# print the accuracy of test data 
print("Accuracy of test data is : " , test_data_accuracy)
```
